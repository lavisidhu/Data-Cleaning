```python
df = pd.read_excel("US Influenza.xlsx")
df.head()
```

```python
# df=df.drop('Age',1) # 0 for row and 1 for column
```

```python
df=df.fillna(0)
# df['Age']=df['AGE 25-64']+df['AGE 25-49']+df['AGE 50-64']
```

```python
df['Age']=df['AGE 25-64']+df['AGE 25-49']+df['AGE 50-64']
```

```python
df=df.drop(['AGE 25-49','AGE 25-64','AGE 50-64'],1)
```

```python
df=df.rename(columns={'Age':'AGE 25-64'})
```

```python
df.head()
```

```python
df=pd.melt(df,id_vars=['YEAR','WEEK','WEIGHTED ILI','UNWEIGHTED ILI','ILITOTAL','TOTAL PATIENTS'],
        value_vars=['AGE 0-4','AGE 5-24','AGE 25-64','AGE 65'],var_name='AGE CATEGORY',value_name='INFLUENZA PATIENTS')
```
